```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import csv
pop2020_df = pd.read_csv("pop2020.csv")
pop2018_df=pd.read_csv("pop2018.csv")
house_df = pd.read_csv("All_house_price.csv")
inventory_df = pd.read_csv("Inventory_sales.csv")
rental_df = pd.read_csv("Rental_price.csv")
```

```python
#sorting out and renaming 2018 population data
pop2018_df=pop2018_df[["Metropolitan areas", "Value"]]
pop2018_df=pop2018_df.rename(columns={"Metropolitan areas":'City', 
                                      "Value":"2018 Population"})

#sorting out and renaming 2020 population data
pop2020_df = pop2020_df.iloc[:, 1:4]
del pop2020_df["State"]
pop2020_df=pop2020_df.rename(columns={"Name" : "City",
                                      "2020 Pop":"2020 Population"})
pop2020_df = pop2020_df.replace({"New York City": "New York"})

#finding common cities
p18_df = pop2018_df["City"]
p20_df = pop2020_df["City"]
p18_df.append(p20_df)
p18_df.drop_duplicates()

#cleaning data based on the common cities
pop2020_df["City"] = p18_df
pop2018_df["City"] = p18_df

#merging the data 
pop_df=pd.merge(pop2018_df, pop2020_df, on="City")
pop_df.to_excel("data.xls")
```

```python
#housing price to df
house_df = house_df.rename(columns={"RegionName":"City"})
house_df.drop(house_df.iloc[:, 3:271], inplace = True, axis = 1) 
house_df.drop(house_df.iloc[:, 0:2], inplace = True, axis = 1) 
house_df['City'] = house_df['City'].str.split(',').str[0]
Cities_df = pop_df["City"]
Phouse_df = pd.merge(Cities_df, house_df, on="City", how="inner")
Phouse_df.head()
```

```python
#rental price to df
rental_df = rental_df.rename(columns={"RegionName":"City"})
rental_df.drop(rental_df.iloc[:, 2:54], inplace = True, axis = 1) 
del rental_df ["RegionID"]
rental_df['City'] = rental_df['City'].str.split(',').str[0]
Cities_df = pop_df["City"]
Prental_df = pd.merge(Cities_df, rental_df, on="City", how="inner")
Prental_df.head()
```

```python
#inventory price to df
inventory_df = inventory_df.rename(columns={"RegionName":"City"})
inventory_df.drop(inventory_df.iloc[:, 3:125], inplace = True, axis = 1) 
inventory_df.drop(inventory_df.iloc[:, 0:2], inplace = True, axis = 1) 
inventory_df['City'] = inventory_df['City'].str.split(',').str[0]
Cities_df = pop_df["City"]
Pinventory_df = pd.merge(Cities_df, inventory_df, on="City", how="inner")
Pinventory_df.head()
```
