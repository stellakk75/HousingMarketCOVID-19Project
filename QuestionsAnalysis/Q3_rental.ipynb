```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
#reading csvs from cleaned data set
cleaned_rental_df = pd.read_csv("../CleanedData/cleaned_rental.csv")
```

```python
del cleaned_rental_df["Unnamed: 0"]
```

```python
#finding the picked cities from the dataframe
df1 = cleaned_rental_df.loc[cleaned_rental_df["City"]=='Boston']
df2 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Salt Lake City"]
df3 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Minneapolis"]
df4 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Seattle"]
df5 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Miami"]
df6 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Atlanta"]
df7 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Columbus"]
df8 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Louisville"]
df9 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Portland"]
df10 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="San Jose"]
df11 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="New York"]
df12 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Oklahoma City"]
df13 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Denver"]
df14 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Kansas City"]
df15 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Los Angeles"]
df16 = cleaned_rental_df.loc[cleaned_rental_df["City"]=="Urban Honolulu"]
```

```python
#combining the found data into a dateframe
frames = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15,df16]
result_df = pd.concat(frames).reset_index(drop=True)
```

```python
#selecting house price data from 2018 to 2020 for each city
Boston = result_df.iloc[0, 2:]
Salt_Lake_City = result_df.iloc[1, 2:]
Minneapolis = result_df.iloc[2, 2:]
Seattle = result_df.iloc[3, 2:]
Miami =  result_df.iloc[4, 2:]
Atlanta = result_df.iloc[5, 2:]
Columbus = result_df.iloc[6, 2:]
Louisville = result_df.iloc[7, 2:]
Portland = result_df.iloc[8, 2:]
SanJose = result_df.iloc[9, 2:]
NewYork = result_df.iloc[10, 2:]
OklahomaCity = result_df.iloc[11, 2:]
Denver = result_df.iloc[12, 2:]
KansasCity = result_df.iloc[13, 2:]
Los_Angeles = result_df.iloc[14, 2:]
Honolulu = result_df.iloc[15, 2:]
```

```python
#plotting the house price of each city from 2018 to 2020
%matplotlib notebook
plt.plot(Boston, label="Boston")
plt.plot(Salt_Lake_City, label="Salt Lake City")
plt.plot(Minneapolis, label="Minneapolis")
plt.plot(Seattle, label="Seattle")
plt.plot(Atlanta, label="Atlanta")
plt.plot(Louisville, label="Louisville")
plt.plot(SanJose, label="SanJose")
plt.plot(NewYork, label="New York")
plt.plot(OklahomaCity, label="Oklahoma City")
plt.plot(Denver, label="Denver")
plt.plot(KansasCity, label="KansasCity")

plt.title("Rental Prices in Cities from 2018 to 2020")
plt.xlabel("Months")
plt.ylabel("Rental Price")
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
#calcuating yearl average of house price of each city
R2018 = ['City', 'State', '2018-01', '2018-02', '2018-03', '2018-04', '2018-05',
       '2018-06', '2018-07', '2018-08', '2018-09', '2018-10', '2018-11',
       '2018-12']
R2018 = result_df[R2018]
R2018["2018 Average"]=R2018.mean(axis=1)

R2019 = ['City', 'State', '2019-01', '2019-02', '2019-03', '2019-04', '2019-05',
       '2019-06', '2019-07', '2019-08', '2019-09', '2019-10', '2019-11','2019-12']
R2019 = result_df[R2019]
R2019["2019 Average"]=R2019.mean(axis=1)

R2020 = ['City', 'State','2020-01', '2020-02', '2020-03', '2020-04', '2020-05',
       '2020-06', '2020-07', '2020-08', '2020-09']
R2020 = result_df[R2020]
R2020["2020 Average"]=R2020.mean(axis=1)

#making dataframe
Ravg_df = R2018[['City', 'State']]
Ravg_df["2018 Average"] = R2018["2018 Average"]
Ravg_df["2019 Average"] = R2019["2019 Average"]
Ravg_df["2020 Average"] = R2020["2020 Average"]
Ravg_df.head()
```

```python
#calculating yearly changing rate for each city

Rrating_df = R2018[['City', 'State']]
Rrating_df["Changing Rate 2018-2019"] = ((Ravg_df["2019 Average"]- Ravg_df["2018 Average"])/Ravg_df["2019 Average"])*100
Rrating_df["Changing Rate 2019-2020"] = ((Ravg_df["2020 Average"]- Ravg_df["2019 Average"])/Ravg_df["2020 Average"])*100
Rrating_df
```

```python
ax = Rrating_df.plot.barh(color=["green","orange"], rot=0)
ax.set_title("Changing Rate in Rental Price")
ax.set_xlabel("City")
ax.set_ylabel("Changing Rate in Rental Price")
x_axis = np.arange(Rrating_df["City"].nunique())
tick_locations = [value for value in x_axis]
plt.yticks(tick_locations, list(Rrating_df["City"]), rotation='horizontal')
plt.show()
```
