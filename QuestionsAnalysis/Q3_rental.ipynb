```python
#calcuating yearl average of house price of each city
R2018 = ['City', 'State', '2018-01', '2018-02', '2018-03', '2018-04', '2018-05',
       '2018-06', '2018-07', '2018-08', '2018-09', '2018-10', '2018-11',
       '2018-12']
R2018 = result_df[R2018]
R2018["2018 Average"]=R2018.mean(axis=1)

R2019 = ['City', 'State', '2019-01', '2019-02', '2019-03', '2019-04', '2019-05',
       '2019-06', '2019-07', '2019-08', '2019-09', '2019-10', '2019-11','2019-12']
R2019 = result_df[R2019]
R2019["2019 Average"]=R2019.mean(axis=1)

R2020 = ['City', 'State','2020-01', '2020-02', '2020-03', '2020-04', '2020-05',
       '2020-06', '2020-07', '2020-08', '2020-09']
R2020 = result_df[R2020]
R2020["2020 Average"]=R2020.mean(axis=1)

#making dataframe
Ravg_df = R2018[['City', 'State']]
Ravg_df["2018 Average"] = R2018["2018 Average"]
Ravg_df["2019 Average"] = R2019["2019 Average"]
Ravg_df["2020 Average"] = R2020["2020 Average"]
Ravg_df.head()
```

```python
#calculating yearly changing rate for each city

Rrating_df = R2018[['City', 'State']]
Rrating_df["Changing Rate 2018-2019"] = ((Ravg_df["2019 Average"]- Ravg_df["2018 Average"])/Ravg_df["2019 Average"])*100
Rrating_df["Changing Rate 2019-2020"] = ((Ravg_df["2020 Average"]- Ravg_df["2019 Average"])/Ravg_df["2020 Average"])*100
Rrating_df
```

```python
ax = Rrating_df.plot.barh(color=["green","orange"], rot=0)
ax.set_title("Changing Rate in Rental Price")
ax.set_xlabel("City")
ax.set_ylabel("Changing Rate in Rental Price")
x_axis = np.arange(Rrating_df["City"].nunique())
tick_locations = [value for value in x_axis]
plt.yticks(tick_locations, list(Rrating_df["City"]), rotation='horizontal')
plt.show()
```
