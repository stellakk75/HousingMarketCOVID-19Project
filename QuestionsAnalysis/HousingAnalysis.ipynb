```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read csv file 
housing2018_df = pd.read_csv("../CleanedData/cleaned_housing2018.csv")
housing2019_df = pd.read_csv("../CleanedData/cleaned_housing2019.csv")
housing2020_df = pd.read_csv("../CleanedData/cleaned_housing2020.csv")
```

```python
#calculating yearly average of the house price
housing2018_df["Yearly Average"] = housing2018_df.mean(axis=1)
order = ['City', 'State', 'Yearly Average', '1/31/18', '2/28/18', '3/31/18', '4/30/18', '5/31/18',
       '6/30/18', '7/31/18', '8/31/18', '9/30/18', '10/31/18', '11/30/18', '12/31/18']

housing2018_df = housing2018_df[order]
```

```python
#finding the top 20 cities with the highest avg house price
housing_2018top = housing2018_df.nlargest(20, 'Yearly Average')
housing_2018top['City'].unique
```

```python
#finding the bottom 20 cities with the highest avg house price

housing_2018bottom = housing2018_df.nsmallest(20, 'Yearly Average')
housing_2018bottom['City'].unique
```

```python
#calculating the yarly average house price in 2019
housing2019_df["Yearly Average"]=housing2019_df.mean(axis=1)
order = ['City', 'State', 'Yearly Average', '1/31/19', '2/28/19', '3/31/19', '4/30/19', '5/31/19',
       '6/30/19', '7/31/19', '8/31/19', '9/30/19', '10/31/19', '11/30/19', '12/31/19']

housing2019_df = housing2019_df[order]
```

```python
#finding the top 20 cities with the highest avg house price
housing_2019top = housing2019_df.nlargest(20, 'Yearly Average')
housing_2019top['City'].unique
```

```python
#finding the bottom 20 cities with the highest avg house price
housing_2019bottom = housing2019_df.nsmallest(20, 'Yearly Average')
housing_2019bottom['City'].unique
```

```python
#calculating the yarly average house price in 2019
housing2020_df["Yearly Average"]=housing2020_df.mean(axis=1)
order = ['City', 'State', 'Yearly Average','1/31/20', '2/29/20', '3/31/20', '4/30/20', '5/31/20',
       '6/30/20', '7/31/20', '8/31/20', '9/30/20']
housing2020_df = housing2020_df[order]
```

```python
#finding the top20 cities with the highest avg house price
housing_2020top = housing2020_df.nlargest(20, 'Yearly Average')
housing_2020top['City'].unique
```

```python
#finding the bottom 20 cities with the highest avg house price
housing_2020bottom = housing2020_df.nsmallest(20, 'Yearly Average')
housing_2020bottom['City'].unique
```

```python
#summary table of top 20 cities 
City = housing_2018top["City"]
Average = housing_2018top["Yearly Average"]
top10_df=pd.DataFrame({"Top 20 in 2018":City, "Yearly Average of Top 20 in 2018" : Average})

top10_df['Top 20 in 2019']= housing_2019top["City"]
top10_df['Yearly Average of Top 20 in 2019']=housing_2019top["Yearly Average"]

top10_df['Top 20 in 2020']= housing_2020top["City"]
top10_df['Yearly Average of Top 20 in 2020']=housing_2020top["Yearly Average"]
top10_df
```

```python
#summary table of bottom 20 cities 

Bcity = housing_2018bottom["City"]
Baverage = housing_2018bottom["Yearly Average"]
bottom10_df=pd.DataFrame({"Bottom 20 in 2018":Bcity, "Yearly Average of Bottom 20 in 2018" : Baverage})


bottom10_df['Bottom 20 in 2019'] = housing_2019bottom["City"]
bottom10_df['Yearly Average of Bottom 10 in 2019'] = housing_2019bottom["Yearly Average"]

bottom10_df['Bottom 20 in 2020'] = housing_2020bottom["City"]
bottom10_df['Yearly Average of Bottom 20 in 2020'] = housing_2020bottom["Yearly Average"]

bottom10_df
```

```python
#calculating yearly changes 

City = housing2018_df["City"]
State = housing2018_df["State"]
Average2018 = housing2018_df['Yearly Average']
Average2019 = housing2019_df['Yearly Average']
Average2020 = housing2020_df['Yearly Average']

summary_df = pd.DataFrame({"City":City,"State":State,
                          "Average2018" : Average2018,
                          "Average2019" : Average2019,
                           "Average2020": Average2020})

summary_df["2018-2019 changing rate"]=''
summary_df["2019-2020 changing rate"]=''


for row in summary_df:
    rate_18 = ((summary_df['Average2019']- summary_df['Average2018'])/summary_df['Average2019'])*100
    summary_df['2018-2019 changing rate'] = rate_18
    
    rate_19 = ((summary_df['Average2020']- summary_df['Average2019'])/summary_df['Average2020'])*100
    summary_df['2019-2020 changing rate'] = rate_19

summary_df["Diff in 2018-2020 changing rate"]=summary_df["2019-2020 changing rate"]-summary_df["2018-2019 changing rate"]
summary_df.head()
summary_df.to_csv("housing_rate.csv")
```
