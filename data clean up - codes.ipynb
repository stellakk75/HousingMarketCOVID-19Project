```python
import pandas as pd
import numpy as np
```

```python
pop2020_df = pd.read_csv("pop2020.csv")
pop20189_df = pd.read_csv("census.csv", encoding="ISO-8859-1")
pop2018_df=pd.read_csv("pop2018.csv")
house_df = pd.read_csv("All_house_price.csv")
inventory_df = pd.read_csv("Inventory_sales.csv")
rental_df = pd.read_csv("Rental_price.csv")
```

```python
pop2018_df=pop2018_df[["Metropolitan areas", "Value"]]
pop2018_df=pop2018_df.rename(columns={"Metropolitan areas":'City', "Value":"2018 Population"})
pop2018_df.head()
```

```python
pop2020_df = pop2020_df.iloc[:, 1:4]
pop2020_df=pop2020_df.rename(columns={"Name" : "City","2020 Pop":"2020 Population"})
pop2020_df.head()
```

```python
pop_df=pd.merge(pop2018_df, pop2020_df, on="City", how='outer')
pop_df=pop_df.dropna(how="any")
column_names = ["State", "City", "2018 Population", "2020 Population"]
pop_df=pop_df.reindex(columns=column_names)
pop_df.head()
```
