```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read csv file 
housing_df = pd.read_csv("cleaned_inventory.csv")
```

```python
housing_df.columns
```

```python
#2018 inventory price data 
columns_pull = ['City', 'State', '2018-01-31','2018-02-28', '2018-03-31', '2018-04-30',
       '2018-05-31', '2018-06-30', '2018-07-31', '2018-08-31', '2018-09-30',
       '2018-10-31', '2018-11-30', '2018-12-31']
housing_2018 = housing_df[columns_pull]

housing_2018
```

```python
#calculating yearly average of the house price
housing_2018["Yearly Average"]=housing_2018.mean(axis=1)
order = ['City', 'State', 'Yearly Average','2018-01-31','2018-02-28','2018-03-31',
         '2018-04-30','2018-05-31','2018-06-30', '2018-07-31', '2018-08-31','2018-09-30',
         '2018-10-31','2018-11-30', '2018-12-31']

housing_2018 = housing_2018[order]
housing_2018
```

```python
#finding the top 10 cities with the highest avg house price
housing_2018top = housing_2018.nlargest(11, 'Yearly Average')
housing_2018top['City'].unique
```

```python
housing_2018top
```

```python
#finding the bottom 10 cities with the highest avg house price

housing_2018bottom = housing_2018.nsmallest(11, 'Yearly Average')
housing_2018bottom['City'].unique
```

```python
housing_2018bottom
```

```python
#calculating yearly changes 

City = housing_df["City"]
State = housing_df["State"]
Average2018 = housing_2018['Yearly Average']
Average2019 = housing_2019['Yearly Average']
Average2020 = housing_2020['Yearly Average']

summary_df = pd.DataFrame({"City":City,"State":State,
                          "Average2018" : Average2018,
                          "Average2019" : Average2019,
                           "Average2020": Average2020})

summary_df["2018-2019 changing rate"]=''
summary_df["2019-2020 changing rate"]=''

for row in summary_df:
    rate_18 = ((summary_df['Average2019']- summary_df['Average2018'])/summary_df['Average2019'])*100
    summary_df['2018-2019 changing rate'] = rate_18
    rate_19 = ((summary_df['Average2020']- summary_df['Average2019'])/summary_df['Average2020'])*100
    summary_df['2019-2020 changing rate'] = rate_19
    
summary_df.to_csv("housing_rate.csv")
```
