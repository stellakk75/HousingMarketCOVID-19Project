```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rental_df = pd.read_csv("cleaned_rental.csv")
```

```python
#2018 rental price data
col2018 = ['City', 'State', '2018-01','2018-02', '2018-03', '2018-04',
       '2018-05', '2018-06', '2018-07', '2018-08', '2018-09', '2018-10',
       '2018-11', '2018-12']
rental2018 = rental_df[col2018]

#calculating yearly average of the rental price
rental2018["Yearly Average"]=rental2018.mean(axis=1)
order = ['City', 'State',"Yearly Average", '2018-01','2018-02', '2018-03', '2018-04',
       '2018-05', '2018-06', '2018-07', '2018-08', '2018-09', '2018-10',
       '2018-11', '2018-12']

rental2018 = rental2018[order]

#finding the top 10 cities with the highest avg rental price
rental2018top = rental2018.nlargest(11, 'Yearly Average')


#finding the bottom 10 cities with the highest avg rental price
rental2018bottom = rental2018.nsmallest(11, 'Yearly Average')
```

```python
#top10 in 2018
rental2018top['City'].unique
```

```python
#bottom 10 in 2018
rental2018bottom['City'].unique
```

```python
#2019 rental price data
col2019 = ['City', 'State', '2019-01', '2019-02', '2019-03',
       '2019-04', '2019-05', '2019-06', '2019-07', '2019-08', '2019-09',
       '2019-10', '2019-11', '2019-12']
rental2019 = rental_df[col2019]

#calculating the yarly average rental price in 2019
rental2019["Yearly Average"]=rental2019.mean(axis=1)
order = ['City', 'State','Yearly Average', '2019-01', '2019-02', '2019-03',
       '2019-04', '2019-05', '2019-06', '2019-07', '2019-08', '2019-09',
       '2019-10', '2019-11', '2019-12']

rental2019 = rental2019[order]

#finding the top 10 cities with the highest avg rental price
rental2019top = rental2019.nlargest(11, 'Yearly Average')


#finding the bottom 10 cities with the highest avg rental price
rental2019bottom = rental2019.nsmallest(11, 'Yearly Average')
```

```python
#top10 in 2019
rental2019top['City'].unique
```

```python
#bottom 10 in 2019
rental2019bottom['City'].unique
```

```python
#2020 rental price data
col2020 = ['City', 'State', '2020-01', '2020-02', '2020-03',
       '2020-04', '2020-05', '2020-06', '2020-07', '2020-08']
rental2020 = rental_df[col2020]

#calculating the yarly average rental price in 2020
rental2020["Yearly Average"]=rental2020.mean(axis=1)
order = ['City', 'State', 'Yearly Average','2020-01', '2020-02', '2020-03',
       '2020-04', '2020-05', '2020-06', '2020-07', '2020-08']

rental2020 = rental2020[order]

#finding the top 10 cities with the highest avg rental price
rental2020top = rental2020.nlargest(11, 'Yearly Average')


#finding the bottom 10 cities with the highest avg rental price
rental2020bottom = rental2020.nsmallest(11, 'Yearly Average')
```

```python
#top10 in 2020
rental2020top['City'].unique
```

```python
#bottom 10 in 2020
rental2020bottom['City'].unique
```

```python
#summary table of top 10 cities 

City = rental2018top["City"]
Average = rental2018top["Yearly Average"]
top10_df=pd.DataFrame({"Top 10 in 2018":City, "Yearly Average of Top 10 in 2018" : Average})

top10_df['Top 10 in 2019']= rental2019top["City"]
top10_df['Yearly Average of Top 10 in 2019']=rental2019top["Yearly Average"]

top10_df['Top 10 in 2020']= rental2020top["City"]
top10_df['Yearly Average of Top 10 in 2020']=rental2020top["Yearly Average"]
top10_df.to_csv("top10rental.csv")
top10_df
```

```python
#summary table of bottom 10 cities 

Bcity = rental2018bottom["City"]
Baverage = rental2018bottom["Yearly Average"]
bottom10_df=pd.DataFrame({"Bottom 10 in 2018":Bcity, "Yearly Average of Bottom 10 in 2018" : Baverage})


bottom10_df['Bottom 10 in 2019'] = rental2019bottom["City"]
bottom10_df['Yearly Average of Bottom 10 in 2019'] = rental2019bottom["Yearly Average"]

bottom10_df['Bottom 10 in 2020'] = rental2020bottom["City"]
bottom10_df['Yearly Average of Bottom 10 in 2020'] = rental2020bottom["Yearly Average"]
bottom10_df.to_csv("bottom10rental.csv")
bottom10_df
```

```python
#calculating yearly changes 

City = rental_df["City"]
State = rental_df["State"]
Average2018 = rental2018['Yearly Average']
Average2019 = rental2019['Yearly Average']
Average2020 = rental2020['Yearly Average']

summary_df = pd.DataFrame({"City":City,"State":State,
                          "Average2018" : Average2018,
                          "Average2019" : Average2019,
                           "Average2020": Average2020})

summary_df["2018-2019 changing rate"]=''
summary_df["2019-2020 changing rate"]=''

for row in summary_df:
    rate_18 = ((summary_df['Average2019']- summary_df['Average2018'])/summary_df['Average2019'])*100
    summary_df['2018-2019 changing rate'] = rate_18
    rate_19 = ((summary_df['Average2020']- summary_df['Average2019'])/summary_df['Average2020'])*100
    summary_df['2019-2020 changing rate'] = rate_19
    
summary_df.to_csv("rental_rate.csv")
```
